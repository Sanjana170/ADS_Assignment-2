**Ingest and manipulate the data using pandas dataframes. Your program should
include a function which takes a filename as argument, reads a dataframe in World bank format and returns two dataframes: one with years as columns and one with countries as columns. Do not forget to clean the transposed dataframe**

```python
names_PowElc =list(PowElc.columns)
### function to create a transposed dataframe
def transfer_data(PowElc):
    Tra_PowElc = PowElc.transpose()

    Tra_PowElc = PowElcPopula_PP.concatenate((Tra_PowElc, PowElcPopula_PP.vstack(names_PowElc)), axis = 1)
    Tra_PowElc = PowElcPopula_PDS.DataFrame(Tra_PowElc)
    Tra_PowElc.insert(0, "year", names_PowElc)
    country=Tra_PowElc[0:1].values.tolist()
    Tra_PowElc.columns =country
    Tra_PowElc=Tra_PowElc.drop(Tra_PowElc.index[0])
    Tra_PowElc=Tra_PowElc.rename(columns={'Country_Name': 'year'})
    Tra_PowElc=Tra_PowElc.reset_index(drop=True)

    # saving transposed data
    Tra_PowElc.to_csv("Tra_PowElc.csv", index=False)
    print('years as columns', PowElc.shape)
    print('countries as columns', Tra_PowElc.shape)
    return (PowElc, Tra_PowElc)
```

```python
### pearson correlation for Iceland country
from scipy.stats import pearsonr as pnr

Corre, p_val= pnr(Iceland_PowElc, Iceland_POP_GRW)
print("The Pearson Correlation value for Iceland country ", Corre, " with a P-value of =", p_val)
```

```python
All_countries_power_consum= []
cl_PowElc=list(PowElc.columns)
for h in cl_PowElc:
    over_all_power=0
    for f in list(PowElc.index.values):
        pow =PowElc[h][f]
        over_all_power=over_all_power+pow
    All_countries_power_consum.append(over_all_power)
All_countries_power_consum
```

```python
All_countries_power_consum = PowElcPopula_PDS.DataFrame(All_countries_power_consum)
All_countries_power_consum.index = list(PowElc.columns)
All_countries_power_consum = All_countries_power_consum.transpose()
All_countries_power_consum.index = ['sum']
All_countries_power_consum
```
